```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, img_as_float32
from skimage.color import rgb2gray
```

### Load an image

```python
rgbImg = io.imread('data/7x5_unique_colors.png')
plt.imshow(rgbImg)
```

```python
print(type(rgbImg)) # array type
print(rgbImg.shape) # (height, width, channels)
print(rgbImg.size) # total number of elements = H * W * C
print(rgbImg.dtype) # data type

plt.imshow(rgbImg[:,:,0], cmap="gray") # red channel
# plt.imshow(rgbImg[:,:,1], cmap=plt.cm.gray) # green 
# plt.imshow(rgbImg[:,:,2], cmap=plt.cm.gray) # blue
```

### Convert image colors and format

```python
grayImg = rgb2gray(rgbImg) # convert to gray scale
floatImg = img_as_float32(grayImg) # convert to float-32
plt.imshow(floatImg, cmap="gray")
```
