```python
import matplotlib.pyplot as plt
import numpy as np
import skimage
```

```python
# load image
image = skimage.io.imread("assignment1_photo.jpg")

#plt.imshow(image)
#plt.show()
```

```python
# conver gray scale
image_gray = skimage.color.rgb2gray(image)

plt.imshow(image_gray, cmap=plt.cm.gray)
plt.show()
```

```python
# print matrix value (3x5 of top left - [:5, :3])
print(image_gray[:5, :3])
```

```python
# print value at x = 1, y = 2 or [1, 0]
print(image_gray[1, 0])
```
